```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("exercise")
sns.set_style(style = "darkgrid")
sns.barplot(x= "time", y = "pulse",data = data)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("exercise")
sns.set_style(style = "ticks")
sns.barplot(x= "time", y = "pulse",data = data)
plt.show()
```
